```python
import numpy as np
import matplotlib.pyplot as plt
```

The following functions are used to transfer a time series vector to a [Gradian Angular Field(GAF)](https://arxiv.org/pdf/1506.00327v1) notation matrix

I tried to use only fundamental packages such as Numpy

Given a time series $X = \begin{bmatrix} x_1, x_2, \dots, x_n \end{bmatrix}$
Rescale X so that all values fall in the interval [-1,1] or [0,1]:

```python
def zero_one_general(X):
    """
    the input array should be a 1-D Numpy array or vector,
    the output will be a rescaled [0,1] 1-D Numpy array or vector
    """
    output = (X - min(X))/(max(X)-min(X))
    return output

def one_one_general(X):
    """
    the input array should be a 1-D Numpy array or vector,
    the output will be a rescaled [-1,1] 1-D Numpy array or vector
    """
    output = (2*X - min(X)-max(X))/(max(X)-min(X))
    return output  
```

We can represent the rescaled time series $\hat X$ in polar coordinates by encoding the value as angular cosine and the time stamp over the len of vector 

```python
def polar_notation(X):
    """
    the input array should be a 1-D Numpy array or vector,
    the output will be two vector, cosine and radius vector
    """
    Phi = np.arccos(X)
    r = np.linspace(1,len(X),len(X))/len(X)
    return Phi, r
```

```python
X = np.array([-2,231,-34,334,234,2,3,1,100,31,2])
a = one_one_general(X)
Phi, r = polar_notation(a)
```

```python
ax = plt.subplot(111,projection='polar')
ax.plot(Phi,r)
print(Phi)
print(r)
```
